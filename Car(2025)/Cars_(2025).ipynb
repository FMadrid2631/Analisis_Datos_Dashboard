## Limpieza de datos textuales

### Subtask:
Normalizar formatos y corregir inconsistencias en columnas de texto.


**Reasoning**:
Examine unique values in relevant columns to identify inconsistencies.



```python
print("Unique values in 'marca':", df['marca'].unique()[:50])
print("\nUnique values in 'nombre_vehiculo':", df['nombre_vehiculo'].unique()[:50])
print("\nUnique values in 'motor':", df['motor'].unique()[:50])
print("\nUnique values in 'bateria':", df['bateria'].unique()[:50])
print("\nUnique values in 'velocidad_max':", df['velocidad_max'].unique()[:50])
print("\nUnique values in 'performance':", df['performance'].unique()[:50])
print("\nUnique values in 'torque':", df['torque'].unique()[:50])
```

**Reasoning**:
Clean the text columns by converting 'marca' to lowercase, removing units and ranges from 'velocidad_max', 'performance', and 'torque', removing currency symbols, commas, and ranges from 'precio', and stripping leading/trailing spaces from all relevant columns.



```python
# 2. Convert 'marca' to lowercase
df['marca'] = df['marca'].str.lower()

# 3. Clean 'velocidad_max', 'performance', and 'torque'
# Function to clean numeric columns with potential ranges and units
def clean_numeric_column(column):
    # Remove units and extra spaces
    column = column.astype(str).str.replace(r'[^\d.-]+', '', regex=True).str.strip()
    # Handle ranges by taking the minimum value
    column = column.apply(lambda x: x.split('-')[0] if '-' in x else x)
    # Convert to numeric, coercing errors to NaN (will be handled later if needed)
    return pd.to_numeric(column, errors='coerce')

df['velocidad_max'] = clean_numeric_column(df['velocidad_max'])
df['performance'] = clean_numeric_column(df['performance'])
df['torque'] = clean_numeric_column(df['torque'])

# 4. Clean 'precio'
def clean_price_column(column):
    # Remove '$', ',', and spaces
    column = column.astype(str).str.replace(r'[$, ]', '', regex=True).str.strip()
    # Handle ranges by taking the minimum value
    column = column.apply(lambda x: x.split('-')[0] if '-' in x else x)
    # Convert to numeric, coercing errors to NaN
    return pd.to_numeric(column, errors='coerce')

df['precio'] = clean_price_column(df['precio'])

# 5. Strip leading/trailing spaces from other text columns
text_cols = ['nombre_vehiculo', 'motor', 'bateria', 'combustible', 'asientos']
for col in text_cols:
    df[col] = df[col].astype(str).str.strip()

# Display the first few rows and info to check changes
display(df.head())
df.info()
```

## Análisis de outliers

### Subtask:
Análisis de outliers


**Reasoning**:
Identify the numerical columns, calculate quartiles and IQR, define outlier limits, and identify outliers for each numerical column.



```python
# Identify numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Calculate Q1, Q3, and IQR for each numerical column
Q1 = df[numerical_cols].quantile(0.25)
Q3 = df[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

# Define outlier limits
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = {}
for col in numerical_cols:
    col_outliers = df[(df[col] < lower_bound[col]) | (df[col] > upper_bound[col])]
    outliers[col] = col_outliers

# Print the number of outliers per numerical column
print("Número de outliers por columna numérica:")
for col, outlier_df in outliers.items():
    print(f"{col}: {len(outlier_df)} outliers")

# Display the outliers for each column (optional, can be large)
# for col, outlier_df in outliers.items():
#     if not outlier_df.empty:
#         print(f"\nOutliers en la columna '{col}':")
#         display(outlier_df)
```

**Reasoning**:
Analyze the identified outliers and implement a strategy to handle them based on their nature and quantity. Given the nature of car data (luxury cars, high-performance vehicles), outliers in performance, speed, and price might represent valid extreme values rather than errors. For torque, the high number of outliers with extremely large values suggests potential data entry issues or unit inconsistencies, which need further investigation before deciding on a handling strategy. For now, we will keep the outliers as they might represent valuable information about high-end cars, except for potentially erroneous torque values which require closer inspection. However, since the next step is documentation, we will only analyze and report the strategy without modifying the dataframe yet, as per best practices to document before applying changes. We will also need to investigate the torque outliers further in the next step, but for this subtask, we just identify and analyze.



```python
# Analyze the nature and quantity of outliers for each numerical column
print("\nAnálisis de Outliers:")

for col, outlier_df in outliers.items():
    print(f"\nColumna '{col}':")
    print(f"Cantidad de outliers: {len(outlier_df)}")

    if not outlier_df.empty:
        print("Estadísticas descriptivas de los outliers:")
        display(outlier_df[numerical_cols].describe())

        # Business justification for handling strategy
        if col in ['velocidad_max', 'performance', 'precio']:
            print(f"Justificación de negocio para '{col}': Los outliers en esta columna (velocidad máxima, rendimiento, precio) a menudo representan vehículos de alto rendimiento, deportivos o de lujo, que son valores válidos y extremos dentro del mercado automotriz. Eliminarlos o modificarlos podría distorsionar el análisis del segmento de gama alta. Por lo tanto, se decide MANTENER estos outliers para preservar la diversidad y el rango real de los datos del mercado.")
        elif col == 'torque':
             print(f"Justificación de negocio para '{col}': La gran cantidad y magnitud de algunos outliers en la columna de torque sugieren posibles errores de entrada de datos o inconsistencias en las unidades (ej. Nm vs otras unidades, o valores erróneamente grandes). Se necesita una investigación más profunda de estos valores específicos antes de decidir una estrategia. Por ahora, se MANTIENEN los outliers para su posterior revisión, pero se señala la necesidad de validación de datos para estos valores extremos.")
```

## Validación de datos

### Subtask:
Realizar verificaciones lógicas en las columnas para asegurar su consistencia.


**Reasoning**:
Perform logical checks on the specified columns to ensure data consistency and identify any invalid entries as per the instructions.



**Reasoning**:
The previous code failed because it tried to convert NaN values in the temporary 'asientos_numeric' column to integers. I need to modify the condition to handle NaN values correctly by filtering them out before checking for positive integers.



**Reasoning**:
The error persists because the `.apply(lambda x: x == int(x))` part of the condition is still being applied to all rows, including those where `asientos_numeric` is NaN, even though they are filtered out in the final selection. I need to apply the integer check only to the non-NaN values.



```python
import numpy as np
import pandas as pd

# Ensure numerical columns are correctly converted, handling errors
def clean_and_convert_numeric(column_series):
    # Remove non-numeric characters except for '.' and '-'
    cleaned_series = column_series.astype(str).str.replace(r'[^\d.\-]+', '', regex=True)
    # Handle ranges by taking the first number if a range exists
    cleaned_series = cleaned_series.apply(lambda x: x.split('-')[0] if '-' in x else x)
    # Convert to numeric, coercing errors to NaN
    return pd.to_numeric(cleaned_series, errors='coerce')

df['caballo_fuerza'] = clean_and_convert_numeric(df['caballo_fuerza'])
df['velocidad_max'] = clean_and_convert_numeric(df['velocidad_max'])
df['performance'] = clean_and_convert_numeric(df['performance'])
df['precio'] = clean_and_convert_numeric(df['precio'])
df['torque'] = clean_and_convert_numeric(df['torque'])


# 1. Verificar si hay valores en la columna `asientos` que no sean números enteros positivos.
# Convert 'asientos' to numeric first, coercing errors to NaN
df['asientos_numeric'] = pd.to_numeric(df['asientos'], errors='coerce')

# Identify rows where 'asientos_numeric' is NaN or not a positive integer
invalid_seats = df[~((df['asientos_numeric'].notna()) & (df['asientos_numeric'] > 0) & (df['asientos_numeric'] == np.floor(df['asientos_numeric'])))]


print("Inconsistencias en la columna 'asientos' (no números enteros positivos o NaN):")
display(invalid_seats[['asientos', 'asientos_numeric']])

# Drop the temporary numeric column
df.drop(columns=['asientos_numeric'], inplace=True)


# 2. Verificar si hay valores no numéricos o negativos en `caballo_fuerza`.
# 'caballo_fuerza' should now be numeric, check for negative values.
invalid_horsepower = df[df['caballo_fuerza'] < 0]
print("\nInconsistencias en la columna 'caballo_fuerza' (valores negativos):")
display(invalid_horsepower[['caballo_fuerza']])


# 3. Verificar si hay valores negativos en `velocidad_max`, `performance`, `precio`, y `torque`.
negative_values = {}
for col in ['velocidad_max', 'performance', 'precio', 'torque']:
    negative_values[col] = df[df[col] < 0]
    print(f"\nInconsistencias en la columna '{col}' (valores negativos):")
    display(negative_values[col][[col]])

# 4. Verificar si la columna `combustible` contiene valores que no estén en la lista esperada de tipos de combustible.
# Define the expected list of fuel types (based on the mapping used earlier)
expected_fuel_types = [
    'híbrido enchufable', 'gasolina', 'diesel', 'hibrido', 'electrico',
    'gasolina/diesel', 'gasolina AWD', 'gasolina/hibrido', 'hidrogeno',
    'diesel/gasolina', 'gasolina/EV', 'hibrido/electrico', 'hibrido (gasolina)',
    'CNG/gasolina', 'hibrido/gasolina', 'gasolina (hibrido)',
    'hibrido (gas + electrico)', 'gas / hibrido'
]
unexpected_fuel_types = df[~df['combustible'].isin(expected_fuel_types)]
print("\nInconsistencias en la columna 'combustible' (valores inesperados):")
display(unexpected_fuel_types[['combustible']].drop_duplicates())
```

**Reasoning**:
The error indicates a TypeError because the 'caballo_fuerza' column is still being treated as a string (`object` dtype) despite the cleaning steps. I need to explicitly convert 'caballo_fuerza' to a numeric type before checking for negative values. The same applies to 'velocidad_max', 'performance', 'precio', and 'torque', although the error specifically pointed to 'caballo_fuerza'. I will ensure all relevant numerical columns are correctly converted to numeric types before performing the checks.

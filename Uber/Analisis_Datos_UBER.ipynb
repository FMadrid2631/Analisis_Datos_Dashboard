## Gestión de valores faltantes

### Sub Tarea:
Identificar las columnas con valores nulos y decidir la estrategia para manejarlos (eliminación, imputación, etc.), justificando la decisión desde una perspectiva de negocio.


**Rasonamiento**:
Identificar las columnas con valores nulos y la cantidad de nulos en cada una, así como el porcentaje de valores nulos para entender la magnitud del problema.



```python
null_counts = df.isnull().sum()
null_percentages = (null_counts / len(df)) * 100

null_info = pd.DataFrame({
    'Null Count': null_counts,
    'Null Percentage': null_percentages
})

print("Información de valores nulos por columna:")
print(null_info[null_info['Null Count'] > 0])
```

**Rasonamiento**:
Para cada columna con valores nulos, determinar la estrategia de manejo más adecuada considerando su tipo de dato y el posible impacto en el análisis del dashboard. Justificar la estrategia elegida desde la perspectiva de un analista de negocio e implementar la estrategia de manejo de valores nulos elegida para cada columna.



```python
# Columnas con alto porcentaje de nulos (más del 80%):
# 'Cancelled Rides by Customer' (93%), 'Reason for cancelling by Customer' (93%)
# 'Cancelled Rides by Driver' (82%), 'Driver Cancellation Reason' (82%)
# 'Incomplete Rides' (94%), 'Incomplete Rides Reason' (94%)
# Estas columnas tienen muy pocos datos útiles para un análisis significativo de las razones o la cantidad de cancelaciones/incompletos directamente.
# Desde una perspectiva de negocio, estas columnas con tantos valores nulos no aportan valor directo para un dashboard general de análisis de viajes.
# Sin embargo, las columnas de 'Reason' podrían ser útiles si se analiza solo el subconjunto de datos con valores no nulos, pero dado el bajo volumen, es probable que no sean representativas.
# Para el propósito de un dashboard general, imputar o intentar derivar valores para el 90%+ de datos faltantes sería especulativo y podría sesgar el análisis.
# La eliminación de estas columnas es la estrategia más limpia para este análisis general, ya que su información es escasa y no fundamental para métricas clave como el valor de la reserva, la distancia o las calificaciones.

df = df.drop(columns=[
    'Cancelled Rides by Customer', 'Reason for cancelling by Customer',
    'Cancelled Rides by Driver', 'Driver Cancellation Reason',
    'Incomplete Rides', 'Incomplete Rides Reason'
])

# Columnas con porcentaje moderado de nulos (entre 30% y 40%):
# 'Avg CTAT' (32%), 'Booking Value' (32%), 'Ride Distance' (32%),
# 'Driver Ratings' (38%), 'Customer Rating' (38%), 'Payment Method' (32%)
# Estas columnas son métricas clave para entender el rendimiento de los viajes, el valor generado y la satisfacción de usuarios y conductores.
# Eliminar filas con nulos en estas columnas reduciría significativamente el dataset (aproximadamente el 30-40% de las filas), perdiendo información valiosa.
# La imputación es una mejor estrategia. El método de imputación debe considerar el tipo de dato y la naturaleza de la métrica.
# Para métricas numéricas continuas como 'Avg CTAT', 'Booking Value', 'Ride Distance', 'Driver Ratings', y 'Customer Rating', la mediana es a menudo una opción robusta contra valores atípicos.
# Para una columna categórica como 'Payment Method', imputar con la moda (el valor más frecuente) es una estrategia común.

# Imputar columnas numéricas con la mediana (Mediana: Es el valor que ocupa la posición central en un conjunto de datos ordenados de menor a mayor.)
numeric_cols_to_impute = ['Avg CTAT', 'Booking Value', 'Ride Distance', 'Driver Ratings', 'Customer Rating']
for col in numeric_cols_to_impute:
    median_value = df[col].median()
    df[col].fillna(median_value, inplace=True)

# Imputar columna categórica con la moda (Moda: Es el valor que más veces se repite dentro de un conjunto de datos.)
if not df['Payment Method'].mode().empty:
    mode_value = df['Payment Method'].mode()[0]
    df['Payment Method'].fillna(mode_value, inplace=True)
else:
    print("La columna 'Payment Method' no tiene valores válidos para imputar con la moda.")
    # Maneja el caso donde el modo está vacío (todos los valores son NaN o la columna está vacía)
    # Para este conjunto de datos, la moda no estará vacío para el Método de pago, pero esta es una buena práctica.
    pass


# Columna con bajo porcentaje de nulos (7%):
# 'Avg VTAT' (7%)
# Esta columna representa el tiempo promedio de llegada del vehículo. Aunque tiene menos nulos, sigue siendo una métrica de tiempo continua.
# Imputarla con la mediana es una estrategia adecuada, similar a las otras columnas numéricas.

median_vtat = df['Avg VTAT'].median()
df['Avg VTAT'].fillna(median_vtat, inplace=True)

# Verificar que los valores nulos hayan sido tratados correctamente
print("\nVerificación de valores nulos después del tratamiento:")
print(df.isnull().sum())
```

Las columnas "Estado de la reserva", "Tipo de vehículo", "Método de pago" y "Día" se muestran consistentes después de la limpieza inicial y no requieren más manipulación de texto para esta subtarea.

## Análisis de outliers

### Sub Tarea:
Identificar posibles valores atípicos en columnas numéricas (`Avg VTAT`, `Avg CTAT`, `Booking Value`, `Ride Distance`, `Driver Ratings`, `Customer Rating`) y determinar cómo manejarlos si es necesario.


**Rasonamiento**:
Identificar las columnas numéricas y calcular estadísticas descriptivas, incluidos los cuartiles, para comprender su distribución e identificar posibles valores atípicos.



```python
numeric_cols = ['Avg VTAT', 'Avg CTAT', 'Booking Value', 'Ride Distance', 'Driver Ratings', 'Customer Rating']

print("Estadísticas descriptivas para columnas numéricas:")
display(df[numeric_cols].describe())
```

**Rasonamiento**:
Calcule el rango intercuartil (RIC) y los límites inferior y superior para la detección de valores atípicos utilizando la regla 1,5\*RIC para cada columna numérica.



```python
# Calculate IQR for each numeric column
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Calculate outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("\nIQR, Límite inferior y límite superior para la detección de valores atípicos:")
outlier_bounds = pd.DataFrame({
    'Q1': Q1,
    'Q3': Q3,
    'IQR': IQR,
    'Lower Bound (Q1 - 1.5*IQR)': lower_bound,
    'Upper Bound (Q3 + 1.5*IQR)': upper_bound
})
display(outlier_bounds)
```

**Rasonamiento**:
Conte el número de valores atípicos en cada columna numérica según los límites inferior y superior calculados. A continuación, considere el contexto empresarial para analizar la naturaleza de estos valores atípicos y la estrategia para gestionarlos sin modificar el marco de datos en esta etapa.



```python
# Identify outliers for each column
outliers_count = {}
for col in numeric_cols:
    col_outliers = df[(df[col] < lower_bound[col]) | (df[col] > upper_bound[col])]
    outliers_count[col] = len(col_outliers)

print("\nNúmero de outliers por columna (usando 1.5*IQR):")
for col, count in outliers_count.items():
    print(f"- {col}: {count}")

print("\nAnálisis de outliers y estrategia propuesta (Sin modificar el DataFrame en esta etapa):")

print("\nAvg VTAT:")
print(f"  Límite Inferior: {lower_bound['Avg VTAT']:.3f}, Límite Superior: {upper_bound['Avg VTAT']:.3f}, Outliers: {outliers_count['Avg VTAT']}")
print("  Los valores atípicos pueden representar tiempos de llegada del vehículo excepcionalmente rápidos o lentos. Los valores negativos del límite inferior no son físicamente posibles para el tiempo, lo que indica que no hay outliers por debajo del rango plausible. Los outliers superiores (por encima de 19.1 minutos) podrían indicar problemas de tráfico, distancia inusual o retrasos del conductor.")
print("  Estrategia: Para un dashboard general, estos valores extremos pueden ser importantes para identificar problemas operativos. Se recomienda mantener estos datos para visualizar la distribución completa y posiblemente investigar las causas de los tiempos de llegada muy altos. No se eliminarán ni modificarán en esta etapa.")

print("\nAvg CTAT:")
print(f"  Límite Inferior: {lower_bound['Avg CTAT']:.3f}, Límite Superior: {upper_bound['Avg CTAT']:.3f}, Outliers: {outliers_count['Avg CTAT']}")
print("  Los outliers (por debajo de 13.15 y por encima de 44.75 minutos) en el Tiempo Promedio de Llegada del Cliente pueden indicar problemas similares a Avg VTAT pero desde la perspectiva del cliente. Valores muy bajos podrían ser errores de datos o viajes excepcionalmente rápidos (si CTAT incluye tiempo de viaje), mientras que valores muy altos indican retrasos significativos para el cliente.")
print("  Estrategia: Al igual que con Avg VTAT, mantener estos outliers es valioso para entender la experiencia del cliente y la eficiencia operativa en casos extremos. Se conservarán los datos brutos.")

print("\nBooking Value:")
print(f"  Límite Inferior: {lower_bound['Booking Value']:.3f}, Límite Superior: {upper_bound['Booking Value']:.3f}, Outliers: {outliers_count['Booking Value']}")
print("  Los valores de reserva por encima de 822.88 son considerados outliers. Esto podría representar viajes de muy larga distancia o tarifas especiales. Un límite inferior negativo o muy bajo (17.88) no es inusual ya que los valores de reserva comienzan en 50, indicando que no hay outliers de bajo valor usando esta regla.")
print("  Estrategia: Los valores de reserva altos son financieramente relevantes y no necesariamente errores. Representan ingresos atípicamente altos por viaje. Se recomienda mantener estos datos para un análisis completo de los ingresos. No se modificarán.")

print("\nRide Distance:")
print(f"  Límite Inferior: {lower_bound['Ride Distance']:.3f}, Límite Superior: {upper_bound['Ride Distance']:.3f}, Outliers: {outliers_count['Ride Distance']}")
print("  Las distancias de viaje por encima de 50.585 km se identifican como outliers. Un límite inferior negativo (-2.575) confirma que no hay outliers de distancia muy corta según esta regla (el mínimo es 1 km). Distancias muy largas son plausibles para viajes de Uber y no necesariamente errores.")
print("  Estrategia: Las distancias de viaje largas son información válida y crucial para entender los patrones de uso y la segmentación de viajes. Se deben mantener estos datos para análisis de distancia y valor de viaje. No se modificarán.")

print("\nDriver Ratings:")
print(f"  Límite Inferior: {lower_bound['Driver Ratings']:.3f}, Límite Superior: {upper_bound['Driver Ratings']:.3f}, Outliers: {outliers_count['Driver Ratings']}")
print("  Las calificaciones de conductor fuera del rango [4.050, 4.450] son consideradas outliers. Dado que las calificaciones son generalmente entre 1 y 5, los valores atípicos identificados (principalmente 5.0 y quizás algunos 3.0-3.5) representan calificaciones excepcionalmente buenas o, menos comúnmente, bajas. El número de outliers bajos (si los hay) es pequeño, mientras que hay un número significativo de 5.0.")
print("  Estrategia: Las calificaciones son cruciales para evaluar el rendimiento del conductor. Las calificaciones de 5.0 son datos válidos y deseados. Las calificaciones bajas (si existen) también son importantes para identificar conductores con bajo rendimiento. Se recomienda mantener todos los valores de calificación ya que representan la opinión real del usuario. No se modificarán.")

print("\nCustomer Rating:")
print(f"  Límite Inferior: {lower_bound['Customer Rating']:.3f}, Límite Superior: {upper_bound['Customer Rating']:.3f}, Outliers: {outliers_count['Customer Rating']}")
print("  Las calificaciones de cliente fuera del rango [3.850, 5.050] se identifican como outliers. Similar a las calificaciones del conductor, esto principalmente captura calificaciones de 5.0. El límite superior de 5.050 indica que los valores de 5.0 no son detectados como outliers superiores por esta regla, mientras que los valores por debajo de 3.850 sí lo son.")
print("  Estrategia: Las calificaciones de cliente (principalmente por debajo de 3.850) son importantes para identificar usuarios problemáticos. Mantener estas calificaciones es relevante para el análisis de comportamiento del usuario. Las calificaciones de 5.0 son datos válidos y se deben mantener. No se modificarán.")

print("\nConclusión sobre el manejo de outliers:")
print("Basado en el análisis y la naturaleza de las métricas, los valores identificados como outliers (usando la regla 1.5*IQR) en las columnas numéricas clave ('Avg VTAT', 'Avg CTAT', 'Booking Value', 'Ride Distance', 'Driver Ratings', 'Customer Rating') parecen representar eventos reales y relevantes para el negocio (ej. viajes largos, reservas de alto valor, calificaciones perfectas/bajas, tiempos de espera/viaje extremos). Eliminar o modificar estos valores podría sesgar el análisis y ocultar información importante para la toma de decisiones.")
print("Por lo tanto, la estrategia propuesta es **mantener los outliers identificados** en el DataFrame original para la creación del dashboard. Esto permitirá visualizar la distribución completa de los datos y analizar el impacto de estos eventos extremos.")
```
